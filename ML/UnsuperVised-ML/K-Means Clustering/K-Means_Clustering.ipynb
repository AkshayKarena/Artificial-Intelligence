```python
features = ['track_popularity','danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms']
x = df[features]
x.head()
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
```

```python
x_scaler.shape
```

```python
x_scaler[:5]
```

```python
kmeans_random = KMeans(n_clusters = 3, random_state = 42)
random_clusters = kmeans_random.fit_predict(x_scaler)
```

```python
print(f"WCSS for K=3 {kmeans_random.inertia_}")
```

```python
plt.scatter(x_scaler[:,2], x_scaler[:,3], c = random_clusters)
plt.xlabel("Energy")
plt.ylabel("Loudness")
plt.title("Clusters on Energy vs Loudness")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
x_pca = pca.fit_transform(x_scaler)


plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c = random_clusters)
plt.title("Clusters with K=3 (Random Choice)")
plt.xlabel("PCA - 1")
plt.ylabel("PCA - 2")
plt.show()
```

```python
WCSS = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, random_state = 42)
    kmeans.fit_predict(x_scaler)
    WCSS.append(kmeans.inertia_)

plt.figure(figsize = (8,6))
plt.plot(range(1,11), WCSS, marker = 'o')
plt.title("Elbow Method")
plt.xlabel("Numbers of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

for k in range(2,11):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    labels = kmeans.fit_predict(x_scaler)
    print(f"{k}, {silhouette_score(x_scaler, labels)}")
```

```python
optimal_k = KMeans(n_clusters= 6, random_state = 42)
optimal_clusters = optimal_k.fit_predict(x_scaler)
print(f"WCSS for K = 6:-{optimal_k.inertia_}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c = optimal_clusters)
plt.title("Clusters with K=6 (Optimal Number Find using Elbow)")
plt.xlabel("PCA - 1")
plt.ylabel("PCA - 2")
plt.show()
```

```python
silhouette_k = KMeans(n_clusters =2, random_state=42)
cluster = silhouette_k.fit_predict(x_scaler)
print(f"WCSS for K = 2:-{silhouette_k.inertia_}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c = cluster)
plt.title("Clusters with K=2 (Optimal Number Find using Silhouette Method)")
plt.xlabel("PCA - 1")
plt.ylabel("PCA - 2")
plt.show()
```
