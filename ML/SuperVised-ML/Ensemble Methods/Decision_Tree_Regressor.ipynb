```python
df_encoded = pd.get_dummies(df, columns = ["sex", "smoker", "region"])
```

```python
X = df.drop("charges",axis = 1)
Y = df[["charges"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

```python
# Now we apply OneHotEncode In x_train in column Sex, Smoker and Region
categorical_cols = ["sex", "smoker", "region"]
ct = ColumnTransformer(transformers=[("onehot", OneHotEncoder(drop = "first"),categorical_cols)],remainder = "passthrough")
```

```python
x_train_encoded = ct.fit_transform(x_train)
```

```python
x_test_encoded = ct.transform(x_test)
```

```python
x_train_encoded[:5]
```

```python
x_test_encoded[:5]
```

```python
dt = DecisionTreeRegressor(random_state = 42, ccp_alpha = 0.01, min_samples_leaf = 20)
dt.fit(x_train_encoded, y_train)
```

```python
y_pred = dt.predict(x_test_encoded)
y_pred[:10]
```

```python
mse= mean_squared_error(y_test, y_pred)
mse
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
plt.figure(figsize=(20,10))
plot_tree(dt, feature_names = ct.get_feature_names_out(), filled = True)
plt.show()
```
